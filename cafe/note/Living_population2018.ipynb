```r
# 컬럼 이름 저장해놓기
colTemp <- colnames(Living_population_201701_01[,3:ncol(Living_population_201701_01)])
colTemp
```

```r
# 자치구코드 기준으로 합치기
Living_population_201701_02 <- aggregate(Living_population_201701_01[,3] ~ 자치구코드, Living_population_201701_01, sum)

for(i in 4:ncol(Living_population_201701_01)){
    
    Living_Temp <- 
        aggregate(Living_population_201701_01[,i] ~ 자치구코드, Living_population_201701_01, sum)
    
    Living_population_201701_02 <- 
        cbind(Living_population_201701_02, Living_Temp)
}


# cbind(
# aggregate(Living_population_201701_01[,3] ~ 자치구코드, Living_population_201701_01, sum),
# aggregate(Living_population_201701_01[,4] ~ 자치구코드, Living_population_201701_01, sum),
# aggregate(Living_population_201701_01[,5] ~ 자치구코드, Living_population_201701_01, sum))

# merge(Living_population_201701_02,
#       aggregate(Living_population_201701_01[,5] ~ 자치구코드, Living_population_201701_01, sum)
#      )

Living_population_201701_02
```

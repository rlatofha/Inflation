```r
# 수출물가지수 불러오기
Export_Target <- read.csv("../Data/Export_Items_Index.csv")
head(Export_Target,10)
```

```r
# 수입물가지수 불러오기
Import_Target <- read.csv("../Data/Import_Items_Index.csv")
head(Import_Target)
```

```r
# 기간을 연/월로 나누기
year <- data.frame('년'=substr(Export_Target[1:nrow(Export_Target),1],1,4))
```

```r
data.frame('월'=c(1:12))
(nrow(Export_Target)-10)/12
```

```r
month <- c(rep(c(1:12),50),1:10)
#  for(i in 1:((nrow(Export_Target)-10)/12)){
#     temp <- c(1:12)
#     temp <- c(temp, temp)
#   }
# data.frame(c(1:12))
#str(Export_Target)
# rep(c(1:12),50)
month <- data.frame('월'=month)
```

```r
period <- cbind(year, month)
```

```r
Export_Target_period <- cbind(period, Export_Target[,-1])
head(Export_Target_period)
```

```r
# Export_Target column 이름 수정하기
colnames(Export_Target_period) <- column_Exp
colnames(Export_Target_period)
```

```r
kimchi_original <- read.csv("../Data/kimchi_price.csv")
```

```r
str(kimchi_original)
```

```r
plot(kimchi_original[,-c(1:2)])
```

```r
cor(kimchi_original[,-c(1:2)])
```

```r
model <- lm(배추가격 ~ ., data=kimchi_original)
```

```r
model
```

```r
# Model 검토
summary(model)
```

```r
# 예측력을 높이기 위해 상대적으로 기여도가 낮은 변수를 제거
reduced <- step(model, direction = 'backward')
```

```r
summary(reduced)
```

Adjusted R-squared(예측력): 0.6501 = 65%

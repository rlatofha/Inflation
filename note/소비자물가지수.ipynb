```r
CPI_Original <- read.csv("../Data/소비자물가지수.csv")
```

```r
# head, tail 확인하기
head(CPI_Original)
tail(CPI_Original, 20)
```

```r
# na 확인
sum(is.na(CPI_Original))
```

```r
# str 확인
str(CPI_Original)
```

```r
# 필요없는 row 정리
# head(CPI_Original, 6)
CPI_Modify_01 <- 
CPI_Original[-c(2:4,688:nrow(CPI_Original)),]
```

```r
# head, tail 재확인
head(CPI_Modify_01)
tail(CPI_Modify_01)
```

```r
# str 확인: as.character로 변환
str(CPI_Modify_01)
```

```r
# str 확인: as.character로 변환
for(i in 1:ncol(CPI_Modify_01)){
    CPI_Modify_01[,i] <- as.character(CPI_Modify_01[,i])
}
head(CPI_Modify_01)
```

```r
# str 확인: 공백 제거하기
str(CPI_Modify_01)
```

```r
# str 확인: 공백 제거하기
for(i in 1:ncol(CPI_Modify_01)){
    CPI_Modify_01[,i] <- trimws(CPI_Modify_01[,i], which = "both")
}

str(CPI_Modify_01)
```

```r
# colunm 이름 변경하기
colnames(CPI_Modify_01) <-
c("기간", CPI_Modify_01[1,2:ncol(CPI_Modify_01)])
colnames(CPI_Modify_01)
```

```r
# row 정리
CPI_Modify_01 <- CPI_Modify_01[-1,]
head(CPI_Modify_01)
nrow(CPI_Modify_01)
```

```r
for(i in 2:ncol(CPI_Modify_01)){
    cat(colnames(CPI_Modify_01)[i],table(CPI_Modify_01[,i]),'\n')
}
```

```r
CPI_Modify_02 <- CPI_Modify_01[120:nrow(CPI_Modify_01),]
```

```r
rownames(CPI_Modify_02) <- c()
rownames(CPI_Modify_02)
```

```r
head(CPI_Modify_02)
tail(CPI_Modify_02)
```

```r
# CPI DATA 기간을 연/월로 나누기
year <- data.frame('년' = substr(CPI_Modify_02[1:(nrow(CPI_Modify_02)-11),1],1,4))
head(year); tail(year,13)
nrow(CPI_Modify_02);nrow(year)
```

```r
# CPI DATA 기간을 연/월로 나누기
(nrow(CPI_Modify_02)-11)/12
month <- c(rep(c(1:12),46))
month <- data.frame('월'=month)
head(month)
nrow(month)
```

```r
# CPI DATA 기간을 연/월로 나누기: 프레임 합치기
period <- cbind(year, month)
head(period); tail(period)
```

```r
# CPI DATA 기간을 연/월로 나누기: 기간 데이터 교체하기
CPI_period <- 
cbind(period, CPI_Modify_02[1:552,-1])
# head(CABdata_period)
```

```r
str(CPI_period)
```

```r
write.csv(CPI_period, "../Data/CPI(소비자물가지수, 품목별).csv", row.names = F)
```

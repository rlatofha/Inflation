```r
# 수출물가지수 불어오기
Export_feature <- read.csv("../Data/Export_Price.csv")
```

```r
# 수입물가지수 불러오기
Import_feature <- read.csv("../Data/Import_Price_Index.csv")
```

```r
# 기간을 연/월로 나누기
year <- data.frame('년'=substr(Export_feature[1:nrow(Export_feature),1],1,4))
nrow(year)
```

```r
# 기간을 연/월로 나누기
(nrow(Export_feature)-10)/12
month <- c(rep(c(1:12),50),1:10)
month <- data.frame('월'=month)
head(month)
nrow(month)
```

```r
# 연월 데이터 프레임 합치기
period <- cbind(year, month)
head(period)
```

```r
# 기간 데이터 교체하기
Export_feature_period <- 
cbind(period, Export_feature[,-1])
head(Export_feature_period)
```

```r
colnames(Export_feature_period)
```

```r
# Export_feature column 이름 수정하기
column_Exp <- 
c('년', '월', paste(colnames(Export_feature[,2:ncol(Export_feature)]),'_수출'))
column_Exp
```

```r
# Export_Target column 이름 수정하기: 공백제거하기
library(stringr)
```

```r
# Export_Target column 이름 수정하기: 공백제거하기
column_Exp <- str_replace_all(column_Exp, ' ', '')
column_Exp
```

```r
# Export_feature column 이름 수정하기
colnames(Export_feature_period) <- column_Exp
colnames(Export_feature_period)
head(Export_feature_period)
```
